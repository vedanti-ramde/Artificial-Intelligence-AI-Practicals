```python
def solve_cryptarithmetic(equation):
    # Split equation into operands and result
    # Format expected: "SEND + MORE = MONEY"
    left, right = equation.replace(" ", "").split('=')
    addends = left.split('+')
    unique_letters = list(set("".join(addends) + right))
    if len(unique_letters) > 10:
        return "Too many unique letters (max 10)."
    # Identify leading letters to prevent assigning them to zero
    leading_letters = {word[0] for word in addends + [right] if len(word) > 1}
    mapping = {}
    used_digits = [False] * 10
    def is_valid(mapping):
        # Calculate the sum of addends
        total_sum = 0
        for word in addends:
            val = int("".join(str(mapping[c]) for c in word))
            total_sum += val
        # Calculate the result value
        result_val = int("".join(str(mapping[c]) for c in right))
        return total_sum == result_val
    def backtrack(idx):
        # Base case: all letters assigned
        if idx == len(unique_letters):
            return mapping if is_valid(mapping) else None
        char = unique_letters[idx]
        for digit in range(10):
            # Check constraints: digit not used and no leading zero
            if not used_digits[digit]:
                if digit == 0 and char in leading_letters:
                    continue
                # Assign and recurse
                mapping[char] = digit
                used_digits[digit] = True
                result = backtrack(idx + 1)
                if result: return result
                # Backtrack: undo assignment
                used_digits[digit] = False
                del mapping[char]
        return None
    solution = backtrack(0)
    return solution
# Example Usage
puzzle = "SEND + MORE = MONEY"
sol = solve_cryptarithmetic(puzzle)
if sol:
    print(f"Solution for {puzzle}: {sol}")
else:
    print("No solution found.")
```
